# Solving an asset price model with heterogeneous beliefs via PEA

In this note I solve an asset pricing model with internal rationality that mimics the setup exposed in Adam, Marcet and Beutel (2017) for the case of heterogeneous beliefs. I do so by using the Parameterized Expectations Approach.
<br>

## Model Structure
The investor's problem can be written as (normalized by dividends, to get stationary ratios):
\begin{equation}
      \max_{\substack{\{S_t^i\}_{t=0}^\infty}} \mathbb{E}_0^{\mathcal{P}_i} \sum_{t=0}^\infty \delta^t \frac{D_t^{1-\gamma}}{1-\gamma} \Bigg(\frac{C_t^i}{D_t}\Bigg)^{1-\gamma}
\end{equation}
s.t. 

\begin{equation}
    \frac{C_t^i}{D_t} + \frac{P_t}{D_t}S_t^i = \Bigg(\frac{P_t}{D_t} + 1 \Bigg)S_{t-1}^i + \frac{W_t}{D_t}
\end{equation}

<br>

\begin{equation}
   \underline{S} \leq S_t^i \leq \bar{S}
\end{equation}
Lower bound: rules out Ponzi schemes; upper bound: aggregate stock supply (equal to the number of agents in the economy). They two ensure a compact feasible set.
<br>
<br>


**Exogenous resource processes**
<br>
1.- Dividends
\begin{equation}
 ln D_t = ln a + ln D_{t-1} + ln \varepsilon_t^D
\end{equation}
<br>

2.- Wage-dividend ratio
\begin{equation}
  ln \Big(1 + \frac{W_t}{D_t}\Big) =  (1-p)ln(1 + \rho) + p ln\Big(1 + \frac{W_{t-1}}{D_{t-1}}\Big) + ln \varepsilon_t^W
\end{equation}
The ln-innovations being jointly Normal, with covariance $\sigma_{DW}$.



**Information system**
<br>
Agents know the processes for dividends and wage-dividends. However, they ignore the exact way these fundamentals map into prices (which are exogenous to them, i.e. they are price-takers). 
<br>
Investors view prices as evolving according to this process

\begin{equation}
 ln P_t = ln b_{t}^i + ln P_{t-1} + ln \varepsilon_t^P
\end{equation}

with the permanent price growth $b_t$ being a drift that vary slowly over time

\begin{equation}
 ln b_{t}^i = ln b_{t-1}^i + ln \nu_t
\end{equation}

The innovations $ln\varepsilon_t^P$ and $ln \nu_t$ being jointly Normal, but uncorrelated. 
<br>

The permanent price growth component $b_t$ is unobserved. It is estimated from the price signals that arises each period via the Kalman filter. Thus, the posterior beliefs are given by

\begin{equation}
   ln b_t^i \sim \mathcal{N} (ln \beta_t^i, \sigma^2)
\end{equation}

where $\sigma^2$ is the steady state variance of the posterior and $\beta_t^i$ is the prior. It is optimally updated (with an information lag) according to 
<br>

\begin{equation}
 ln \beta_t^i = ln \beta_{t-1}^i + g^i (ln P_{t-1} - ln P_{t-2} - ln \beta_{t-1}^i)
\end{equation}

**Source of beliefs heterogeneity**
<br>
It is assumed that agents learn differently (some faster, some slower). In other words, they extrapolate past prices with different intensities. There is some survey evidence backing that (Adam, Beutel, Marcet and Merkel (2015)). It follows that when there are good (bad) news, the fast learners become more (less) optimistic, given rise to trade. For the application, I assume the existence of just two groups, with $g^1 > g^2$. 
<br>
<br>



## Solution approach

The concavity of the objective function and the convexity set guarantee the sufficiency of the First Order Conditions for an interior optimal plan.

**Optimality conditions**
<br>
1.- Euler Equation:
\begin{equation}
    (CD_t^i)^{-\gamma} = \delta \mathbb{E}_t^\mathcal{P}\Bigg(\Big(\frac{D_{t+1}}{D_t}\Big)^{1-\gamma}\frac{(PD_{t+1}+1)}{PD_t}(CD_{t+1}^i)^{-\gamma}\Bigg)
\end{equation}

2.- Budget Constraint:
\begin{equation}
    \frac{C_t^i}{D_t} + \frac{P_t}{D_t}S_t^i = \Bigg(\frac{P_t}{D_t} + 1 \Bigg)S_{t-1}^i + \frac{W_t}{D_t}
\end{equation}

3.- Market clearing condition
\begin{equation}
 \sum_i S_t^i = \sum_i S_{-1}^i \equiv S  (\text{fixed stock supply})
\end{equation}

The approach consists on: 
<br>
 1.-  Solve the Euler Equation for $CD_t^i$. Approximate the RHS via a parametric function of the states. 
<br>
 2.-  Plug the consumption policy on the budget constraint to pin down the stock policy function.
 <br>
 3.-  Use the market clearing condition to determine equilibrium prices. 
<br> 
Let's cover the steps one by one. 
<br>
<br>


**Step 1: Approximating the consumption policy**
<br>
Hakansson (1970, Econometrica) derived closed form solutions for the consumption policy function for a similar model. Assuming $W_t = 0$ for all t, the consumption policy is linear in wealth: 
\begin{equation}
  C_t ^* = BZ_t
\end{equation}
where 
<br>
\begin{equation}
B= 1 - \Big(\delta \mathbb{E}_t(u(R_{t+1}))(1-\gamma)\Big)^{1/\gamma}
\end{equation}
with $R_t = (P_t + D_t) / P_{t-1}$. 
<br>
Notice that under Rational Expectations, $\mathbb{E}_t(u(R_{t+1}))$ is just a constant, so that B (i.e., the propensity to consume out of wealth) is just a constant. Besides, $Z_t$ is the wealth at the beginning of the period: $Z_t = (P_t + D_t)S_{t-1}$. 
<br>
<br>

This solution does not hold in the present model because: i) there is stochastic wage income; ii) under learning, the expected utility of stock returns is non-constant. Under these circumstances, there is no way of getting a closed form solution since we cannot open up $\Bigg(\frac{W_{t+1}}{D_{t+1}} + R_{t+1}\Big(\frac{Z_t}{D_t} - \frac{C_t}{D_t}\Big)\Bigg)^{1-\gamma}$ (the Binomial Theorem cannot be applied here: $1-\gamma$ can be both negative and not an integer). 
<br>



*----------------------------
<br>
Note: if one foregoes stochastic wage income, the model can be solved even under learning quite easily. If investors learn directly about stock returns, $\mathbb{E}_t^{\mathcal{P}}(u(R_{t+1}))=x_t$ can be directly computed, in which case B would become a time varying object $B_t$. Thus, time-varying stock return beliefs would change the propensity to consume/invest out of wealth. (This alternative is not very appealing to me because it does not allow to include capital gains and transaction taxes, which is what I am after).
<br>
-----------------------------* 
<br>
<br>


*----------------------------
<br>
Note: to save notation here, I'm writing $XD_t$ standing for $X_t/D_t$.
<br>
-----------------------------* 


However, a consumption policy linear in wealth is appealing because it would allow to pin down equilibrium prices analytically from the market clearing condition. This is the approach I follow. 
<br>


The function that approximates the consumption policy is

\begin{equation}
CD_t^i = CD (S_{t-1}^i, PD_t, WD_t, \pi_t^i) = B(\pi_t^i)\Big(WD_t + (PD_t + 1)S_{t-1}^i\Big)
\end{equation}

with $B(\pi_t^i)$ having some resemblance to Hakansson's closed form solution (*This is not needed plus Hakansson's formula is not a solution for the current model. However, it turned out to perform very well*). 

\begin{equation}
  B_t^i = B(\pi_t^i) = 1 - \Big(\delta \chi\pi_t^i (1-\gamma)\Big)^{1/\gamma}
\end{equation}

where $\pi_t^i$ stands for return expectations and $\chi$ is an unknown parameter which will be estimated via PEA (see below). Particularly,

\begin{equation}
 \pi_t^i = \mathbb{E}_t^{\mathcal{P}_i} \Big(\frac{P_{t+1} + D_{t+1}}{P_t}\Big) = \mathbb{E}_t^{\mathcal{P}_i} \Big(\frac{P_{t+1}}{P_t}\Big) + \mathbb{E}_t^{\mathcal{P}_i} \Big(\frac{D_{t+1}}{D_t}\Big)\frac{D_t}{P_t} = \beta_t^i + a\frac{D_t}{P_t}
\end{equation}
<br>

Intuition for the consumption policy: it is linear in wealth; the propensity to consume depends (negatively) on stock return expectations. 


**Step 2: Getting the stock policy function**
<br>
Then, plugging the consumption policy in the budget constraint, we can pin down the stock policy function:
\begin{equation}
 S_t^i = (1 - B_t^i)\frac{\Big(WD_t + (PD_t + 1)S_{t-1}^i\Big)}{PD_t}
\end{equation}
<br>


**Step 3: Determining market-clearing prices**
<br>
Finally, given that policies are linear in wealth, I can determine analytically the PD ratio that clears the market. For the two agents example, the market clearing condition reads as 
\begin{equation}
 S_t^1 + S_t^2 = \Big((1 - B_t^1)(WD_t + (PD_t + 1)S_{t-1}^1)/PD_t\Big) + \Big((1 - B_t^2)(WD_t + (PD_t + 1)S_{t-1}^2)/PD_t\Big) = \bar{S}
\end{equation}
Therefore, solving for prices
\begin{equation}
 PD_t = \frac{WD_t(2 - B_t^1 - B_t^2) + (1 - B_t^1)S_{t-1}^1 + (1-B_t^2)S_{t-1}^2}{\bar{S} - (1 - B_t^1)S_{t-1}^1 - (1-B_t^2)S_{t-1}^2}
\end{equation}

The denominator is always positive as long as $\bar{S} > (1 - B_t^1)S_{t-1}^1 - (1-B_t^2)S_{t-1}^2$ or $B_t^1S_{t-1}^1 + B_t^2S_{t-1}^2>0$. A sufficient condition for that is $B_t^i>0$ and $S_{t-1}^i>0$. In the first application, I will ensure that by imposing a projection facility (however, in the simulation this PF is never used).
<br>
<br>


*----------------------------*
### PEA approach to estimate the policy function parameter. 
The only remaining task is to establish the value of $\chi$. For that, I follow PEA. 
<br>
The Euler Equation can be rewritten as
\begin{equation}
(CD_t^i)^{-\gamma}\delta^{-1} = \mathbb{E}_t^\mathcal{P}\Bigg(\Big(\frac{D_{t+1}}{D_t}\Big)^{1-\gamma}\frac{(PD_{t+1}+1)}{PD_t}(CD_{t+1}^i)^{-\gamma}\Bigg)
\end{equation}

To apply PEA, I follow the next steps:
<br>

1.- Substitute $CD_t^i$ by the approximation function, which becomes dependent on $\chi$. 
<br>

2.- Compute the interior of the RHS expectation, according to investor's beliefs. Since investors know the process for dividends and wage-dividends, the only problematic objects are $PD_{t+1}$ and $\beta_{t+1}$. In this case:

\begin{equation}
PD_{t+1} = \frac{P_{t+1}}{P_t}\frac{D_{t}}{D_{t+1}}PD_t
\end{equation}
According to agents' view: 
\begin{equation}
\frac{P_{t+1}}{P_t} = \beta_{t+1}\varepsilon_{t+1}^p
\end{equation}
and
\begin{equation}
\beta_{t+1}=\beta_t\nu_{t+1}
\end{equation}
<br>

3.- Minimize the Euler Equation square residuals

$$
G(\chi) = \underset{\chi}{\operatorname{argmin}}\Bigg\{\Bigg(\Big(\frac{D_{t+1}^{\mathcal{P}}}{D_t}\Big)^{1-\gamma}\frac{(PD_{t+1}^{\mathcal{P}}+1)}{PD_t}(CD(\chi)_{t+1}^{i,\mathcal{P}})^{-\gamma}\Bigg)-\frac{(CD(\chi)_t^i)^{-\gamma}}{\delta}\Bigg\}^2
$$


The superscript $\mathcal{P}$ highlights that those variables are computed according to investors' view. 
<br>

4.- Find a fixed point $\chi = G(\chi)$. For that, update $\chi$ following

\begin{equation}
 \chi^{j+1} = \chi^j + d(G(\chi^j) - \chi^j)
\end{equation}

where j iteration number and d the "dampening" parameter. 
<br>
<br>

-----------------------------
## Code

First, specify parameters and the exogenous income processes.

```python
#-------------------------------------------
"""Packages"""
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as optimize
import warnings
warnings.filterwarnings('ignore')




#--------------------------------------------
"""Parameters"""

N=4000 #Periods
delta=0.99 #Discount rate
a=1.001 #Mean dividend growth
s_d = 0.0167 #Std. deviation div. growth
s_w= 0.0197 #Std. deviation wage-dividend growth
s_dw=-0.0000374 #Covariance(wage-dividend,dividend)
p=0.98 #Persistence wage-dividend process.
rhow=22 #1+rhow is the average consumption-dividend ratio. 
gamma = 2 #Risk aversion parameter
sigma_e=s_d #Std. deviation transitory component price growth
sigma_v=s_d/5 #Std. deviation unobserved permanent component price growth innovation 

#Derived parameters
rho=np.exp((gamma/2)*(pow(s_w,2)*(1+gamma) - (1-gamma)*pow(s_d,2)))*np.exp(-gamma*(1-gamma)*s_dw)   #E(e_d[t+1]^{1-gamma} e_w[t+1]^{-gamma})        
pi=np.exp(-gamma*(1-gamma)*pow(s_d,2)/2) #E(e_d[t+1]^{1-gamma})
landa=delta*pow(a,1-gamma)*rho 
beta_RE_d=pow(a,1-gamma)*rho 
pd_RE=landa/(1-landa)

#Heterogeneous learning gains
gain1=0.04
gain2=0.01

#Aggregate stock supply
supply=2





#------------------------------------------------
"""Exogenous processes"""
np.random.seed(10) #Reproducible simulation

#---- Dividends and wage-dividends -----
mean, cov = [(-pow(s_d,2)/2), (-pow(s_w,2)/2)], [(s_d**2, s_dw), (s_dw, s_w**2)]
shocks = np.random.multivariate_normal(mean, cov, N)
e_d=np.exp(shocks[:,0])
e_w=np.exp(shocks[:,1])


dg=np.ones((N))
wd=np.ones((N))
for t in range(1,N):
    dg[t]=a*e_d[t]
    wd[t]=np.exp((1-p)*np.log(1+rhow) + p*np.log(wd[t-1]) + np.log(e_w[t]))


#------ Subjective price growth ------
#Transitory component
mean, cov = [-(sigma_e**2)/2,sigma_e**2]
eps_p = np.random.normal(mean, cov, N)

#Innovation unobserved component price growth
mean, cov = [-(sigma_e**2)/2,sigma_e**2]
eps_v = np.random.normal(mean, cov, N)
```

2.- Write the function that computes the residuals of the Euler Equation.

```python
#---------------------------------------------------
"""Function that computes the residuals of the Euler Equation.

   Input: parameter chi
   Output: sum of square residuals of EE. 

"""


def residuals(chi):
    
    """Storage objects"""
    
    #Beliefs
    beta1=np.zeros((N))
    beta1[1]=beta_RE_d
    beta2=np.zeros((N))
    beta2[1]=beta_RE_d
    log_beta1=np.zeros((N))
    log_beta2=np.zeros((N))

    #k Hakansson-like
    k1=np.ones((N))
    k2=np.ones((N))
    
    #Propensity to consume out of wealth
    B1=np.ones((N))
    B2=np.ones((N))
    
    #Stocks 
    s1=np.ones((N))*1
    s2=np.ones((N))*1
    
    #Consumption
    c1=np.ones((N))
    c2=np.ones((N))
    
    #Equilibrium prices
    pd=np.ones((N))
    pd[0]=pd_RE
    
    #Price growth vector
    pg=np.ones((N))
    
    #Prediction error
    prediction_error1=np.ones((N))
    prediction_error2=np.ones((N))
    
    #Beliefs dispersion
    disagreement=np.ones((N))
    
    #Transactions
    change_s1=np.ones((N))
    change_s2=np.ones((N))
    
    #Trading Volume (over total stocks)
    volume=np.ones((N))
    
    #Check market clearing
    clearing=np.ones((N))
    goods_clear=np.ones((N))
    

    """Step 1: Compute the time series for a given chi"""
    for t in range(1,N-1):    
        #k Hakansson-like
        k1[t]=chi*(beta1[t]+ a/pd[t-1])
        k2[t]=chi*(beta2[t] +a/pd[t-1])
            
        #Time-varying propensity to consume out of wealth
        B1[t]=1 - pow((delta*k1[t]*(1-gamma)),1/(gamma))
        B2[t]=1 - pow((delta*k2[t]*(1-gamma)),1/(gamma))
        
        #Projection facility
        #Restrict beliefs within a range that produce propensities to consume between 0 and 1 (no borrowing)
        if B1[t]<=0 or B1[t]>=1:
            B1[t]=B1[t-1]
            beta1[t]=beta1[t-1]
            
        if B2[t]<=0 or B2[t]>=1:
            B2[t]=B2[t-1]
            beta2[t]=beta2[t-1]

    
        #Equilibrium PD
        pd[t]=((2 - B1[t] - B2[t])*(wd[t]-1) + (1-B1[t])*s1[t-1] + (1-B2[t])*s2[t-1])/(supply - (1-B1[t])*s1[t-1] - (1-B2[t])*s2[t-1])
        
        #Price growth
        pg[t]=(pd[t]/pd[t-1])*dg[t]
        
        #Prediction error
        prediction_error1[t]= np.log((pg[t]))- np.log(beta1[t])
        prediction_error2[t]= np.log((pg[t]))- np.log(beta2[t])
    
        #Beliefs updating
        log_beta1[t+1]=np.log(beta1[t]) + gain1*prediction_error1[t]
        log_beta2[t+1]=np.log(beta2[t]) + gain2*prediction_error2[t]
    
        beta1[t+1]=np.exp(log_beta1[t+1])
        beta2[t+1]=np.exp(log_beta2[t+1])

        #Disagreement
        disagreement[t]=beta1[t]-beta2[t]
        
        #Effective stock
        s1[t]=((1-B1[t])*(wd[t]-1 + (pd[t]+1)*s1[t-1]))/pd[t]
        s2[t]=((1-B2[t])*(wd[t]-1 + (pd[t]+1)*s2[t-1]))/pd[t]
    
        
        #Transactions
        change_s1[t]=s1[t]-s1[t-1]
        change_s2[t]=s2[t]-s2[t-1]
            
        #Volume traded over total stocks
        volume[t]=(abs(change_s1[t])+abs(change_s2[t]))/supply
        
        #Consumption
        c1[t]=B1[t]*(wd[t]-1 + (pd[t]+1)*s1[t-1])
        c2[t]=B2[t]*(wd[t]-1 + (pd[t]+1)*s2[t-1])
    
        #Market clearing
        clearing[t]=supply - s1[t] - s2[t]
        goods_clear[t]=2*(wd[t]) - c1[t] - c2[t]
        
            
    #-------------------------------
    """Step 2: Compute the interior of the expectation in the Euler Equation"""
        
    #Price growth according the agents view    
    e_pg1=np.ones((N)) 
    e_beta1=np.ones((N))
    for t in range(1,N):
        e_beta1[t]=beta1[t-1]*np.exp(eps_v[t-1])
        e_pg1[t]=e_beta1[t]*np.exp(eps_p[t])
        
    #Simulate the interior of the expectation
    y1=np.ones((N))
    e_pd1=np.ones((N))
    e_B=np.ones((N))
    e_cd=np.ones((N))
    for t in range(N-1):
        #Next period subjective pd ratio
        e_pd1[t+1]=(pd[t]/dg[t+1])*e_pg1[t+1]
        
        #Next period subjective propensity to consume
        e_B[t+1]= 1 - pow((delta*chi*(e_beta1[t+1] +a/pd[t])*(1-gamma)),1/(gamma))
        
        #Next period subjective consumption
        e_cd[t+1]=e_B[t+1]*(wd[t+1]-1 + (e_pd1[t+1]+1)*s1[t])
        
        #Interior of the expectation
        y1[t+1]=pow(dg[t+1],1-gamma)*((e_pd1[t+1]+1)/pd[t])*pow(e_cd[t+1],-gamma)
    
    
    residu=np.zeros((N))
    approx=np.zeros((N))

    for t in range(1,N-1):
        approx[t]=pow(c1[t],-gamma)/delta #From Euler Equation.
        residu[t]=y1[t+1]-approx[t]
        
    ssr=sum(residu[500:-50]**2) #Burn first 500 periods and 50 final periods. 

        
    return ssr
```

3.- Minimize the EE residuals and update the parameter $\chi$ until reaching convergence.

```python
d=0.001
tol=0.01
toler=0.00005
diff=1
count=0
max_number=2000
sresiduals=[]
chi=-0.95
while abs(diff)>tol:
    
    result=optimize.minimize(residuals,chi,tol=toler)
    resi=result.fun
    sresiduals.append(resi)

    #Update coefficients:
    chi_new=d*result.x + (1-d)*chi
    diff=result.x-chi
    chi=chi_new
    
    #Alternative endings of the algorithm
    if count>=1:
        #if sresiduals[count]>sresiduals[count-1]:
           #print('No further reduction in EE residuals')
           #break
        
        if count>max_number:
            print('Max number of iterations reached')            
            break
    
    count +=1
    #print(count)    
    if abs(diff)<=tol:
        print('Convergence achieved. Chi=',chi)
        print('Euler Equation Sum Square Residuals =', resi)
```

- The algorithm got a fixed point.
<br>
- Besides, the Euler Equation residuals turn out to be very small, which means that the approximation works well. 
<br>
- Now that we have the right parameter value, we can simulate the model. 

```python
chi=-0.93912334 #From PEA

#Beliefs
beta1=np.zeros((N))
beta1[1]=beta_RE_d
beta2=np.zeros((N))
beta2[1]=beta_RE_d
log_beta1=np.zeros((N))
log_beta2=np.zeros((N))

#k Hakansson-like
k1=np.ones((N))
k2=np.ones((N))

#Propensity to consume out of wealth
B1=np.ones((N))
B2=np.ones((N))

#Stocks 
s1=np.ones((N))
s2=np.ones((N))

#Consumption
c1=np.ones((N))
c2=np.ones((N))

#Equilibrium prices
pd=np.ones((N))
pd[0]=pd_RE

#Price growth vector
pg=np.ones((N))

#Prediction error
prediction_error1=np.ones((N))
prediction_error2=np.ones((N))

#Beliefs dispersion
disagreement=np.ones((N))

#Transactions
change_s1=np.ones((N))
change_s2=np.ones((N))

#Trading Volume (over total stocks)
volume=np.ones((N))

#Check market clearing
clearing=np.ones((N))
goods_clear=np.ones((N))


"""Step 1: Compute the time series for a given chi"""
for t in range(1,N-1):    
    #k Hakansson-like
    k1[t]=chi*(beta1[t]+ a/pd[t-1])
    k2[t]=chi*(beta2[t] +a/pd[t-1])

    #Time-varying propensity to consume out of wealth
    B1[t]=1 - pow((delta*k1[t]*(1-gamma)),1/(gamma))
    B2[t]=1 - pow((delta*k2[t]*(1-gamma)),1/(gamma))

    #Projection facility
    #Restrict beliefs to ensure positive prices (never used).
    if B1[t]<=0 or B1[t]>=1:
        B1[t]=B1[t-1]
        beta1[t]=beta1[t-1]

    if B2[t]<=0 or B2[t]>=1:
        B2[t]=B2[t-1]
        beta2[t]=beta2[t-1]


    #Equilibrium PD
    pd[t]=((2 - B1[t] - B2[t])*(wd[t]-1) + (1-B1[t])*s1[t-1] + (1-B2[t])*s2[t-1])/(supply - (1-B1[t])*s1[t-1] - (1-B2[t])*s2[t-1])

    #Price growth
    pg[t]=(pd[t]/pd[t-1])*dg[t]

    #Prediction error
    prediction_error1[t]= np.log((pg[t]))- np.log(beta1[t])
    prediction_error2[t]= np.log((pg[t]))- np.log(beta2[t])

    #Beliefs updating
    log_beta1[t+1]=np.log(beta1[t]) + gain1*prediction_error1[t]
    log_beta2[t+1]=np.log(beta2[t]) + gain2*prediction_error2[t]

    beta1[t+1]=np.exp(log_beta1[t+1])
    beta2[t+1]=np.exp(log_beta2[t+1])

    #Disagreement
    disagreement[t]=beta1[t]-beta2[t]

    #Effective stock
    s1[t]=((1-B1[t])*(wd[t]-1 + (pd[t]+1)*s1[t-1]))/pd[t]
    s2[t]=((1-B2[t])*(wd[t]-1 + (pd[t]+1)*s2[t-1]))/pd[t]


    #Transactions
    change_s1[t]=s1[t]-s1[t-1]
    change_s2[t]=s2[t]-s2[t-1]

    #Volume traded over total stocks
    volume[t]=(abs(change_s1[t])+abs(change_s2[t]))/supply

    #Consumption
    c1[t]=B1[t]*(wd[t]-1 + (pd[t]+1)*s1[t-1])
    c2[t]=B2[t]*(wd[t]-1 + (pd[t]+1)*s2[t-1])

    #Market clearing
    clearing[t]=supply - s1[t] - s2[t]
    goods_clear[t]=2*(wd[t]) - c1[t] - c2[t]
```

**Disagreement and trade**
<br>
Altogether, there is trade because there is disagreement. Specifically, it happens that the more (less) optimistic investor buys (sells) to the other group.  This is in line with the classical results first posed by Harrison and Kreps (1978). 
<br>

In the following graph, I plot disagreement (i.e., $\beta_t^1 - \beta_t^2$) against the change in the stocks of investor 1 (i.e., $S_t^1 - S_{t-1}^1$). Their correlation is very high (0.86). 
<br>

Notice that the fact there are points in the 2nd and 4th quadrants sets out that sometimes the more optimistic agent has to sell stocks. It must happen because agent 2 is a bit less optimistic ($\beta_t^2 < \beta_t^1$) but has more assets ($S_{t-1}^2 >> S_{t-1}^1$), overcoming the larger propensity to invest of agent 1. When the disagreement is high enough the more optimistic investor always buy, though. 
<br>

It is a matter of the stock scale. This situation disappears if I increase the aggregate stock supply (for instance, to 2000 instead of 2). In this case, the correlation (disagreement, trade) becomes 1.

**Stock demand function**

The following graphs explore some properties of the stock policy function. 

The first graph plot it in the $\{S_t^i, PD_t\}$ plane, like a standard demand function. It is downwards sloping, ensuring a single equilibrium price. A rise in optimism moves it to the right, increasing the equilibrium price as a result. 

```python
s_function=np.ones((N))
s_function1=np.ones((N))
pd_grid=np.linspace(100,800,N)
B1=0.03
B2=0.025
wda=22
ss=2
for t in range(N):
    s_function[t]=((1-B1)*(wda + (pd_grid[t]+1)*ss))/pd_grid[t]
    s_function1[t]=((1-B2)*(wda + (pd_grid[t]+1)*ss))/pd_grid[t]

plt.plot(s_function, pd_grid, label='Pessimistic')
plt.plot(s_function1, pd_grid, label='Optimistic')
plt.axvline(x=2, color='k', LineStyle='--')
plt.xlabel('stocks')
plt.ylabel('PD ratio')
plt.xlim([1.96,2.05])
plt.text(2.005, 700, 'stock supply')
plt.legend()
```

It can be represented in a 2D plane with any of the others states. In the next graph it is plotted in the $\{S_{t-1}^i, S_t^i\}$ plane, like a policy function.  

As observed, when the investor has little stocks, it is optimal to accumulate more. However, the opposite is true when she is rich enough. That can be viewed as a stabilization mechanism (when poor, save; when rich, consume) (related to wealth effect?) that imparts endogenous mean-reversion to stock holdings and then, to prices.

```python
s_grid=np.linspace(0.6,1.4,N)
s_function=np.ones((N))
s_function1=np.ones((N))
pda=600
B1=0.05
B2=0.0365
wda=22

for t in range(N):
    s_function[t]=((1-B1)*(wda + (pda+1)*s_grid[t]))/pda
    s_function1[t]=((1-B2)*(wda + (pda+1)*s_grid[t]))/pda

#plt.plot(s_grid, s_function, label='pessimistic')
plt.plot(s_grid, s_function1, color='orange', label='optimistic')
plt.plot(s_grid, s_grid,color='Black', linestyle=':',linewidth=2)
plt.xlabel('$S_{t-1}$')
plt.ylabel('$S_{t}$')
plt.xlim([0.9,1.1])
plt.ylim([0.9,1.1])
```
